# Importing libraries

```python
import numpy as np 
import pandas as pd 
import sqlite3 as sql
```

# Importing dataset

```python
conn = sql.connect('movies.sqlite')

print("SQL database connected")

# the number of tables
table = pd.read_sql("SELECT name FROM sqlite_master WHERE type = 'table'", conn)
print(table)
```

# Executing tasks

```python
# get all data about directors
query = """
    SELECT *
    FROM directors;
"""
directors = pd.read_sql(query, conn)
directors
```

```python
# get all data about movies
query = """
    SELECT *
    FROM movies;
"""
movies = pd.read_sql(query, conn)
movies
```

```python
# get all data about directors
query = """
    SELECT *
    FROM sqlite_sequence;
"""
directors = pd.read_sql(query, conn)
directors
```

```python
# total number of movies
query = """
    SELECT count (*)
    FROM movies;
"""
movies_all = pd.read_sql(query, conn)
movies_all
```

```python
#  Find all directors with name starting with John
query = """
    SELECT *
    FROM directors
    Where name like 'John%';
"""
directors_name = pd.read_sql(query, conn)
directors_name
```

```python
# Find these 3 directors: James Cameron ; Luc Besson ; John Woo
query = """
    SELECT *
    FROM directors
    Where name == 'James Cameron' or name=='Luc Besson' or name=='John Woo' ;
"""
directors_names = pd.read_sql(query, conn)
directors_names
```

```python
# Count female directors
query = """
    SELECT count(*)
    FROM directors
    Where gender == '1';
"""
directors_female = pd.read_sql(query, conn)
directors_female
```

```python
# 5 most popular movies
query = """
    SELECT *
    FROM movies
    order by popularity desc
    limit 5;
"""
popular_movies = pd.read_sql(query, conn)
popular_movies
```

```python
#  7 movies with the biggest budget
query = """
    SELECT *
    FROM movies
    order by budget desc
    limit 7;
"""
movies_budget = pd.read_sql(query, conn)
movies_budget
```

```python
#  7 most bankable movies
query = """
    SELECT *
    FROM movies
    order by revenue desc
    limit 7;
"""
movies_bankable = pd.read_sql(query, conn)
movies_bankable
```

```python
# the most awarded average vote since the January 1st, 2000
query = """
    SELECT original_title, vote_average
    FROM movies
    where release_date > '2000-01-01'
    ORDER by vote_average  DESC
    limit 1;
"""
movies_vote = pd.read_sql(query, conn)
movies_vote 
```

```python
# movies were directed by James Cameron
query = """
    SELECT original_title
    FROM movies 
    Join directors 
    ON movies.director_id = directors.id
    where name = 'James Cameron'
"""
movies_cameron = pd.read_sql(query, conn)
movies_cameron
```

```python
# 10 directors made the most movies
query = """
    SELECT name, count(name) as number_of_movies
    FROM movies 
    Join directors 
    ON movies.director_id = directors.id
    group by director_id
    order by count(name) DESC
    limit 10;
"""
movies_all = pd.read_sql(query, conn)
movies_all
```

```python
# top 10 directors have the highest revenue
query = """
    SELECT name, revenue
    FROM movies 
    Join directors 
    ON movies.director_id = directors.id
    group by director_id
    order by revenue DESC
    limit 10;
"""
directors_revenue = pd.read_sql(query, conn)
directors_revenue
```
